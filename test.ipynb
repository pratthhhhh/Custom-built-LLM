# Model Inference Test

Trained the model for 5 epochs due to limited resources. While it might output random words, the model is able the form English chunks. If I train it till convergence, this can perform well.

```python
import sys
sys.path.append(".")  # ensure repo root is on the path

import torch
from model import GPTModel
from config import GPT_CONFIG, DEVICE
from lora import apply_lora
from tokenizer import get_tokenizer, text_to_tensor
```

```python
CHECKPOINT  = "out/finetune/lora/final"
LORA_RANK   = 16
LORA_ALPHA  = 16.0
MAX_TOKENS  = 200
TEMPERATURE = 0.7
TOP_K       = 50
```

```python
model = GPTModel(GPT_CONFIG).to(DEVICE)
model = apply_lora(model, rank=LORA_RANK, alpha=LORA_ALPHA)

checkpoint = torch.load(CHECKPOINT, map_location=DEVICE)
model.load_state_dict(checkpoint)
model.eval()
print(f"Loaded checkpoint from {CHECKPOINT}")
```

```python
def run(prompt: str):
    tokenizer = get_tokenizer()
    idx = text_to_tensor(prompt, tokenizer)
    with torch.no_grad():
        out = model.generate(idx, MAX_TOKENS, TEMPERATURE, TOP_K)
    new_tokens = out[0][idx.shape[1]:].tolist()
    print(f"Prompt : {prompt}")
    print(f"Response: {tokenizer.decode(new_tokens)}")

run("The United States is a country") # prompt
```
